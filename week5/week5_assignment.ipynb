# Question 1&2
Train a Linear Regression model on the training dataset with fit_intercept = False and compute ‘r2_score’ on the test

For splitting the dataset use the following info

The last column of the dataset (Credit_Limit) corresponds to the target variable. Separate the dataset to have the feature matrix as X and the target variable as y. Use random_state = 42 wherever applicable.


Split the dataset into train and test set with a 70:30 ratio and random_state = 42.

```python
# splitting dataset according to the
from sklearn.model_selection import train_test_split

X = data.drop('Credit_Limit', axis=1)
y = data['Credit_Limit'].copy()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression(fit_intercept=False)

lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score

y_pred = lr.predict(X_test)
r2 = r2_score(y_test, y_pred)

print(f"R² on test set: {r2:.3f}")
```

```python
coeff = lr.coef_
max_index = np.argmax(np.abs(coeff))

print(max_index)
```

# Question 9 & 10

Perform Hyperparameter Tuning using GridSearchCV on AdaBoostRegressor. Use random_state = 42. Value of cv = 4. Hyperparameter tuning is to be done over:

- n_estimators as [10,50,100,200,500]

- learning_rate as [0.1,0.5,1,2]

Use the best model obtained to compute the score on the test data

```python
# splitting dataset according to the
from sklearn.model_selection import train_test_split

X = data.drop('Credit_Limit', axis=1)
y = data['Credit_Limit'].copy()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.ensemble import AdaBoostRegressor

abr = AdaBoostRegressor(random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score


param_grid = {
    'n_estimators': [10, 50, 100, 200, 500],
    'learning_rate': [0.1, 0.5, 1, 2]
}

grid_search = GridSearchCV(
    estimator=abr,
    param_grid=param_grid,
    cv=4,
    scoring='r2',
    n_jobs=-1
)

grid_search.fit(X_train, y_train)


best_model = grid_search.best_estimator_

y_pred = best_model.predict(X_test)
r2 = r2_score(y_test, y_pred)

print(f"r2 = {r2:.3}")
```

```python
best_params = grid_search.best_params_

print(best_params)
```
